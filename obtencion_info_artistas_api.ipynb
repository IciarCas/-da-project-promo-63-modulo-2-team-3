```python
import pandas as pd
import json
import requests
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
api_key = os.getenv("api_key")
```

```python
with open("lista_artistas.json", "r", encoding="utf-8") as archivo:
    lista_artistas = json.load(archivo)

print(lista_artistas)
print(len(lista_artistas))
```

```python
datos_artista = []
url = "http://ws.audioscrobbler.com/2.0"

#for artista in ['Chenoa', 'Alaska']:
for artista in lista_artistas: #[:40]:  # entre30 y 35 no hay similares
    llamada = requests.get(
        url,
        params={
            'api_key': api_key,
            'method': 'artist.getinfo',
            'artist': artista,  # ← sin llaves
            'format': 'json'
        }
    )
    datos = llamada.json()

    try:
        similares = [
            sim.get('name') for sim in datos.get('artist').get('similar').get('artist')
        ]
    except (AttributeError, TypeError):
        similares = []


    datos_artista.append({
        'name': datos.get('artist', {}).get('name'),
        'listeners': datos.get('artist', {}).get('stats', {}).get('listeners'),
        'playcount': datos.get('artist', {}).get('stats', {}).get('playcount'),
        # 'similares': [sim.get('name') for sim in datos.get('artist').get('similar').get('artist')],
        'similares': similares
    })
```
