```python
mycursor.execute("USE musicdata_studiocode;") 
mycursor.execute("""
    CREATE TABLE IF NOT EXISTS generos_musicales (
        id_genero INT PRIMARY KEY AUTO_INCREMENT,
        nombre VARCHAR(50) NOT NULL
    );
""")
```

```python
mycursor.execute("USE musicdata_studiocode;") 
mycursor.execute("""
    CREATE TABLE IF NOT EXISTS artistas_similares (
        id_artista_similares INT PRIMARY KEY AUTO_INCREMENT,
        nombre_similares VARCHAR(70) NOT NULL
    );
""")
```

```python
mycursor.execute("USE musicdata_studiocode;") 
mycursor.execute("""
    CREATE TABLE IF NOT EXISTS canciones (
        id_cancion INT AUTO_INCREMENT,
        id_artista INT NOT NULL,
        titulo_cancion VARCHAR(70) NOT NULL,
        id_genero INT NOT NULL,
        año_lanzamiento YEAR,
        PRIMARY KEY (id_cancion, id_artista),
        CONSTRAINT fk_id_artista_artistas FOREIGN KEY (id_artista)
            REFERENCES artistas(id_artista)
            ON DELETE RESTRICT
            ON UPDATE CASCADE,
        CONSTRAINT fk_id_generos_generos FOREIGN KEY (id_genero)
            REFERENCES generos_musicales(id_genero)
            ON DELETE RESTRICT
            ON UPDATE CASCADE
    );
""")
```

```python
mycursor.execute("USE musicdata_studiocode;") 
mycursor.execute("""
    CREATE TABLE IF NOT EXISTS similares (
        id_artista INT NOT NULL,
        id_artista_similares INT NOT NULL,
        PRIMARY KEY (id_artista, id_artista_similares),
        CONSTRAINT fk_id_artista_artistas2 FOREIGN KEY (id_artista)
            REFERENCES artistas(id_artista)
            ON DELETE RESTRICT
            ON UPDATE CASCADE,
        CONSTRAINT fk_id_artista_similar_artistas_similares FOREIGN KEY (id_artista_similares)
            REFERENCES artistas_similares(id_artista_similares)
            ON DELETE RESTRICT
            ON UPDATE CASCADE
    );
""")
```
