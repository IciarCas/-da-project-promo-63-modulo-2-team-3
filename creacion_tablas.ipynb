````python
import pandas as pd
import json
import requests
````

````python
with open("info_artistas.json", "r", encoding="utf-8") as archivo:
    info_artistas = json.load(archivo)

print(info_artistas)
print(len(info_artistas))
````

````python
# Este bloque es para obtener informacion del artista (sin similares) y asignacion de id.
datos_insercion_artistas = []

artista_id = 1
for artista in info_artistas:
    valores = list(artista.values())[:-1]  # quitamos 'similares'
    artista_con_id = [artista_id] + valores
    datos_insercion_artistas.append(artista_con_id)
    artista_id += 1

print(datos_insercion_artistas)
print(len(datos_insercion_artistas))
````

````python
# Este bloque es para obtener solo similares sin repeticion.

#bloque 1: obtener similares sin duplicados.
datos_insercion_similares = []
similares_unicos = set() # el set hace que sea unico el valor

for artista in info_artistas:
    for similar in artista['similares']:
        similares_unicos.add(similar)

#bloque 2: asignacion de id a similares y guardado
similar_id_map = {} # se crea este diccionario para utilizarlo en la 3r bloque
similar_id = 1

for similar in similares_unicos:
    datos_insercion_similares.append([similar_id, similar])
    similar_id_map[similar] = similar_id
    similar_id += 1

print(similar_id_map)
print(datos_insercion_similares)
````

````python
# Este bloque crea la relacion entre artistas y similares mediante sus id
artista_similares = []

artista_id = 1
for artista in info_artistas:
    for similar in artista['similares']:
        artista_similares.append([
            artista_id,
            similar_id_map[similar]
        ])
    artista_id += 1

 
print(artista_similares)
````

````python
datos_insercion_artistas = []
datos_insercion_similares = []
artista_similares = []

artista_id = 1
similar_id = 1
for artista in info_artistas:
     valores = list(artista.values())[:-1]
     artista_con_id = [artista_id] + valores
     datos_insercion_artistas.append(artista_con_id)

     for similar in artista['similares']:
         similar_con_id = [similar_id, similar]
         datos_insercion_similares.append(similar_con_id)

         ids = [artista_id, similar_id]
         artista_similares.append(ids)

         similar_id += 1

     artista_id += 1

print(datos_insercion_artistas)
print(len(datos_insercion_artistas))
print(datos_insercion_similares)
print(len(datos_insercion_similares))
print(artista_similares)
print(len(artista_similares))
````

```sql
CREATE TABLE artistas (
    id INT PRIMARY KEY AUTO_INCREMENT,
    nombre VARCHAR(100) NOT NULL
);

CREATE TABLE similares (
    id INT PRIMARY KEY AUTO_INCREMENT,
    nombre VARCHAR(100) NOT NULL
);

CREATE TABLE artista_similar (
    artista_id INT,
    similar_id INT,
    PRIMARY KEY (artista_id, similar_id),
    FOREIGN KEY (artista_id) REFERENCES artistas(id),
    FOREIGN KEY (similar_id) REFERENCES similares(id)
);
```
